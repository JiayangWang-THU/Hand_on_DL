```python
# 每日一导
import torch
import torch.nn as nn
import torch.nn.functional as F
import math 
```

```python
class selfAtten(nn.Module):
    def __init__(self,dim):
        super().__init__()
        self.W_q=nn.Linear(dim,dim)
        self.W_k=nn.Linear(dim,dim)
        self.W_v=nn.Linear(dim,dim)
        self.W_o=nn.Linear(dim,dim)
    def forward(self,x):
        # x=[B,T,D]
        d = x.shape[-1]
        Q = self.W_q(x)
        K = self.W_k(x)
        V = self.W_v(x)
        
        logits =  Q @ K.transpose(-2,-1)/math.sqrt(d)
        scores = F.softmax(logits,dim=-1)@V
        output = self.W_o(scores)
        return output
```
