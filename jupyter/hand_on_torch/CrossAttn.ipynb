```python
# 每日一导
import torch
import torch.nn as nn
import torch.functional as F
import math
print(torch.__version__)
```

```python
class CrossAttenSingHead(nn.Module):
    def __init__(self,d_model):
        super().__init__()
        self.w_q = nn.Linear(d_model,d_model,bias= True)
        self.w_k = nn.Linear(d_model,d_model,bias= True)
        self.w_v = nn.Linear(d_model,d_model,bias= True)
        self.w_o = nn.Linear(d_model,d_model,bias= True)
    def forward(self,x_dec,x_enc):
        B,Tq,C = x_dec.shape
        _,Tk,_ = x_dec.shape

        Q = self.w_q(x_dec)
        K = self.w_k(x_enc)
        V = self.w_v(x_enc)

        scores = torch.matmul(Q,K.transpose(-2,-1))/ math.sqrt(C)
        
        attn = torch.matmul(F.softmax(scores,-1),V)
        return attn


```

```python
x_dec = torch.randn(2, 4, 8)   # B=2, Tq=4
x_enc = torch.randn(2, 6, 8)   # B=2, Tk=6

model = CrossAttenSingHead(d_model=8)

out = model(x_dec, x_enc)

print("x_dec:", x_dec.shape)
print("x_enc:", x_enc.shape)
print("out:", out.shape)
```
