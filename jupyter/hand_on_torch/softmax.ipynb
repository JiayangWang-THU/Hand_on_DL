# softmax
作用主要是指数归一化，指数放大差距，然后加个T可以控制能量熵
非负，0到1
变可微


```python
# 每日一导
import numpy as np
import torch
```

```python
# 纯np手搓
# 注意这里输入是个向量
def softmax(x,axis=-1):
    # x:(...,n)
    # 首先要找最大
    # 不然容易上溢出
    m = np.max(x,axis=axis,keepdims=True)
    x_shifted = x - m 
    exp_x = np.exp(x_shifted)
    sum_exp = np.sum(exp_x,axis,keepdims=True)
    return exp_x /sum_exp
```

```python
def softmax_torch(x,dim=-1):
    m = torch.max(x,dim=dim,keepdim=True).values
    x = x - m
    return torch.exp(x)/torch.sum(torch.exp(x),dim=dim,keepdim=True)
```
